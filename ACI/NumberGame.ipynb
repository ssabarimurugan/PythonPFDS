```python
import numpy as np
class HumanPlayer:
    def __init__(self, name):
        self.name = name 
        self.chosenNumbers = []
    
    def chooseNumber(self, numbers):
        while True:
            choice = input("Input your selected number: Available Numbers: " + str(numbers))
            if choice.isnumeric() and int(choice) in numbers:
                self.chosenNumbers.append(int(choice))
                print(self.name + " chose " + choice)
                return int(choice)

class Player:
    def __init__(self, name):
        self.name = name 
        self.chosenNumbers = []
    
    def chooseNumber(self, numbers):
        while True:
            idx = np.random.choice(len(numbers))
            choice = numbers[idx]
            if choice in numbers:
                self.chosenNumbers.append(choice)
                print(self.name + " chose " + str(choice))
                return choice

class Game:
    def __init__(self, p1, p2):
        self.maxNumber = 20
        self.p1 = p1
        self.p2 = p2
        self.isEnd = False
        self.availableNumbers = np.arange(1,self.maxNumber,1)

    def winner(self):
        if len(self.availableNumbers) == 0:
            self.isEnd = True
            print(self.p1.name + " sum is " + str(sum(self.p1.chosenNumbers)))
            print(self.p2.name + " sum is " + str(sum(self.p2.chosenNumbers)))
            if sum(self.p1.chosenNumbers) > sum(self.p2.chosenNumbers):
                return 1
            elif sum(self.p1.chosenNumbers) < sum(self.p2.chosenNumbers):
                return 2
            else:
                return 0
            
        return None
    
    def selectNextPlayer(self):
        if sum(self.p1.chosenNumbers) > sum(self.p2.chosenNumbers):
            return 2
        elif sum(self.p1.chosenNumbers) < sum(self.p2.chosenNumbers):
            return 1
    
    def play(self):
        # Player 1
        self.availableNumbers = np.delete(self.availableNumbers, np.where(self.availableNumbers == self.p1.chooseNumber(self.availableNumbers)))
        while not self.isEnd:
            win = self.winner()
            if win is not None:
                if win == 1:
                    print(self.p1.name, "wins!")
                elif win == 2:
                    print(self.p2.name, "wins!")
                else:
                    print("tie!")
                break
            else:
                player = self.selectNextPlayer()
                if player == 1:
                    # Player 1
                    self.availableNumbers = np.delete(self.availableNumbers, np.where(self.availableNumbers == self.p1.chooseNumber(self.availableNumbers)))
                else:
                    # Player 2
                    self.availableNumbers = np.delete(self.availableNumbers, np.where(self.availableNumbers == self.p2.chooseNumber(self.availableNumbers)))
```

```python
# Computer Vs Computer
p1 = Player("Computer1")
p2 = Player("Computer2")

st = Game(p1, p2)
st.play()
```

```python
# Human Vs Computer
p1 = Player("Computer")
p2 = HumanPlayer("Human")

st = Game(p1, p2)
st.play()
```

```python
# Human Vs Human
p1 = HumanPlayer("Human1")
p2 = HumanPlayer("Human2")

st = Game(p1, p2)
st.play()
```
